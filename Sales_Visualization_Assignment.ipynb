```python
# Import Required Libraries
import numpy as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snp
```

```python
# Create Synthetic Dataset
np.random.seed(42)

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

sales = [12000,15000,13500,17000,16000,19000,
         18500,21000,20000,22500,24000,26000]

subjects = ['Math','Science','English','History','Art']
scores = [78,85,72,90,65]

study_hours = np.random.uniform(1,10,100)
exam_scores = study_hours * 8 + np.random.normal(0,5,100)

product_sales = np.random.normal(loc=50, scale=10, size=200)
```

# Task 1 — Line Plot: Monthly Sales Trend

###Purpose: To analyze sales growth over time.

```python

plt.plot(months, sales, color='blue', marker='o')
plt.title("monthly Sales Trend")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```

# Task 2 — Bar Plot: Subject Score Comparison

###Purpose: To compare marks across subjects.

```python

plt.bar(subjects, scores)
plt.title("subject score comparison")
plt.xlabel("subjects")
plt.ylabel("scores")
plt.show()
```

```python
# import Seaborn lbraries
import seaborn as sns
import matplotlib.pyplot as plt
```

# Task 3 — Scatter Plot: Study Hours vs Exam Score

### Purpose: To compare marks across subjects.

###

```python

sns.scatterplot(x=study_hours, y=exam_scores)
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.show()
```

# Task 4 — Box Plot: Product Sales Distribution

### Purpose: To check Data spread + outliers

```python
sns.boxplot(x=product_sales)
plt.title("product sales distribution")
plt.xlabel("product sales")
plt.show()
```
